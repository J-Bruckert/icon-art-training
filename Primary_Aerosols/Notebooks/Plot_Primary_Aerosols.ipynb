```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import xarray as xr
from glob import glob
import cartopy.crs as ccrs
from FUNCTIONS import get_dz, tracer_load
from matplotlib import ticker
```

## Load the data
Insert the correct path to your output directory

```python
# Enter path to output directory
datadir = '/path/to/data/'
ds = xr.open_mfdataset(datadir+'icon-art-allaero_DOM01_ML_00*',autoclose=True)
dz  = get_dz(ds.isel(time=0))
```

## Sellect tracers for dust, seasalt and soot aerosols
Insert the names of the tracers

```python

dust = ds['???'] +ds['???']

seas = ds['???'] + ds['???'] + ds['???'] + ds['???']

soot = ds['???'] + ds['???']
```

```python
# Tracer load of dust 
dust_load = tracer_load(dust,ds['rho'],dz)*1e-6

# Tracer load of seasalt
seas_load = tracer_load(seas,ds['rho'],dz)*1e-6

# Tracer load of soot
soot_load = tracer_load(soot,ds['rho'],dz)*1e-6
```

## Plot Aerosol Column Loadings after 12 h and 24 h

```python
fig1 = plt.figure(figsize=(12,10))

ax1=fig1.add_subplot(3,2,1,projection=ccrs.PlateCarree(central_longitude=0))
plot1=ax1.contourf(ds.lon, ds.lat, dust_load.isel(time=12),levels=np.arange(0,0.51,0.05),extend='max',cmap='Blues', transform = ccrs.PlateCarree())
ax1.coastlines()
plt.colorbar(plot1,orientation='horizontal')
plt.title('Dust after 12 h')

ax1=fig1.add_subplot(3,2,3,projection=ccrs.PlateCarree(central_longitude=0))
plot2=ax1.contourf(ds.lon, ds.lat, seas_load.isel(time=12),levels=np.arange(0,0.051,0.005),extend='max', cmap='Blues', transform = ccrs.PlateCarree())
ax1.coastlines()
plt.colorbar(plot2,orientation='horizontal')
plt.title('Seasalt after 12 h')

ax1=fig1.add_subplot(3,2,5,projection=ccrs.PlateCarree(central_longitude=0))
plot3=ax1.contourf(ds.lon, ds.lat, soot_load.isel(time=12),levels=np.arange(0,0.021,0.002),extend='max', cmap='Blues', transform = ccrs.PlateCarree())
ax1.coastlines()
plt.colorbar(plot3,orientation='horizontal')
plt.title('Soot after 12 h')

ax1=fig1.add_subplot(3,2,2,projection=ccrs.PlateCarree(central_longitude=0))
plot4=ax1.contourf(ds.lon, ds.lat, dust_load.isel(time=24),levels=np.arange(0,0.51,0.05),extend='max', cmap='Blues', transform = ccrs.PlateCarree())
ax1.coastlines()
plt.colorbar(plot4,orientation='horizontal')
plt.title('Dust after 24 h')

ax1=fig1.add_subplot(3,2,4,projection=ccrs.PlateCarree(central_longitude=0))
plot5=ax1.contourf(ds.lon, ds.lat, seas_load.isel(time=24),levels=np.arange(0,0.051,0.005),extend='max', cmap='Blues', transform = ccrs.PlateCarree())
ax1.coastlines()
plt.colorbar(plot5,orientation='horizontal')
plt.title('Seasalt after 24 h')

ax1=fig1.add_subplot(3,2,6,projection=ccrs.PlateCarree(central_longitude=0))
plot6=ax1.contourf(ds.lon, ds.lat, soot_load.isel(time=24),levels=np.arange(0,0.021,0.002),extend='max', cmap='Blues', transform = ccrs.PlateCarree())
ax1.coastlines()
plt.colorbar(plot6,orientation='horizontal')
plt.title('Soot after 24 h')
```

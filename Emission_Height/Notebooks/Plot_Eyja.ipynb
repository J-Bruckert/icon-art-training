```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
from matplotlib import ticker
```

## Plot horizontal distribution of the ash

```python
ds_4km = xr.open_mfdataset('/scratch/k/k??????/TurbDiff_Ex3/???*.nc')
ds_9km = xr.open_mfdataset('/scratch/k/k??????/TurbDiff_Ex3/???*.nc')
print(list(ds_9km.variables))
```

```python
# get model level height
def get_dz(ds):
    dz = -1 * ds.z_ifc.diff('height_2')
    dz = dz.rename({'height_2':'height'})
    dz = dz.assign_coords(height=(dz.height - 1))
    return dz

# Calculate column integrated tracer mass (tracer load)
def tracer_load(tr,rho,dz):
    m_tr = tr * rho * dz
    m_tr = m_tr.sum('height')
    m_tr = m_tr.squeeze()
    return m_tr
```

```python
# read in all ash tracers
ash_load_4km = tracer_load(ds_4km.ash_insol_acc+ds_4km.ash_insol_coa+ds_4km.ash_giant,ds_4km.rho,get_dz(ds_4km))
ash_load_9km = tracer_load(ds_9km.ash_insol_acc+ds_9km.ash_insol_coa+ds_9km.ash_giant,ds_9km.rho,get_dz(ds_9km))

# convert in g/m2
ash_load_4km = ash_load_4km *1e-6
ash_load_9km = ash_load_9km *1e-6
```

```python
# set colorbar levels
#color_ash = np.arange(0.5, 10, 0.5)
color_ash = np.array([5e-3,1e-2,5e-2,1e-1,5e-1,1e0,5e0,1e1])
```

```python
# Experiment 1: Emission at 4km height
crs_proj = ccrs.PlateCarree()

fig1 = plt.figure(figsize=(13,6))
ax1 = plt.subplot(2,2,1,projection=crs_proj)
c=plt.contourf(ds_4km.lon,ds_4km.lat,ash_load_4km[24,:,:],cmap='Blues',transform=crs_proj,locator=ticker.LogLocator(),levels=color_ash,extend='max')
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-35, 110, 25, 75])
plt.title('Ash plume emitted at 4 km after 24 h')

ax1 = plt.subplot(2,2,2,projection=crs_proj)
c=plt.contourf(ds_4km.lon,ds_4km.lat,ash_load_4km[48,:,:],cmap='Blues',transform=crs_proj,locator=ticker.LogLocator(),levels=color_ash,extend='max')
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-35, 110, 25, 75])
plt.title('Ash plume emitted at 4 km after 48 h')

ax1 = plt.subplot(2,2,3,projection=crs_proj)
c=plt.contourf(ds_4km.lon,ds_4km.lat,ash_load_4km[72,:,:],cmap='Blues',transform=crs_proj,locator=ticker.LogLocator(),levels=color_ash,extend='max')
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-35, 110, 25, 75])
plt.title('Ash plume emitted at 4 km after 72 h')

ax1 = plt.subplot(2,2,4,projection=crs_proj)
c=plt.contourf(ds_4km.lon,ds_4km.lat,ash_load_4km[96,:,:],cmap='Blues',transform=crs_proj,locator=ticker.LogLocator(),levels=color_ash,extend='max')
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-35, 110, 25, 75])
plt.title('Ash plume emitted at 4 km after 96 h')

ax2=fig1.add_axes([0.2,0.05,0.6,0.03])
cbar=plt.colorbar(c,cax=ax2,orientation='horizontal')
cbar.set_label('Tracer column loading in g/m²')
```

```python
# Experiment 2: Emission at 9km height
crs_proj = ccrs.PlateCarree()

fig1 = plt.figure(figsize=(13,6))
ax1 = plt.subplot(2,2,1,projection=crs_proj)
c=plt.contourf(ds_9km.lon,ds_9km.lat,ash_load_9km[24,:,:],cmap='Greens',transform=crs_proj,locator=ticker.LogLocator(),levels=color_ash,extend='max')
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-35, 110, 25, 75])
plt.title('Ash plume emitted at 9 km after 24 h')

ax1 = plt.subplot(2,2,2,projection=crs_proj)
c=plt.contourf(ds_9km.lon,ds_9km.lat,ash_load_9km[48,:,:],cmap='Greens',transform=crs_proj,locator=ticker.LogLocator(),levels=color_ash,extend='max')
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-35, 110, 25, 75])
plt.title('Ash plume emitted at 9 km after 48 h')

ax1 = plt.subplot(2,2,3,projection=crs_proj)
c=plt.contourf(ds_9km.lon,ds_9km.lat,ash_load_9km[72,:,:],cmap='Greens',transform=crs_proj,locator=ticker.LogLocator(),levels=color_ash,extend='max')
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-35, 110, 25, 75])
plt.title('Ash plume emitted at 9 km after 72 h')

ax1 = plt.subplot(2,2,4,projection=crs_proj)
c=plt.contourf(ds_9km.lon,ds_9km.lat,ash_load_9km[96,:,:],cmap='Greens',transform=crs_proj,locator=ticker.LogLocator(),levels=color_ash,extend='max')
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-35, 110, 25, 75])
plt.title('Ash plume emitted at 9 km after 96 h')

ax2=fig1.add_axes([0.2,0.05,0.6,0.03])
cbar=plt.colorbar(c,cax=ax2,orientation='horizontal')
cbar.set_label('Tracer column loading in g/m²')
```

## Identify "No Fly" zones

Here: "no fly" zones are defined as areas where a threshold of $2\,$mg/m$^3$ is exceeded at some point of the vertical column

```python
ash_insol_4km = (ds_4km.ash_insol_acc+ds_4km.ash_insol_coa+ds_4km.ash_giant) * ds_4km.rho * 10**(-3) # transform to mg/m^3
ash_insol_9km = (ds_9km.ash_insol_acc+ds_9km.ash_insol_coa+ds_9km.ash_giant) * ds_9km.rho * 10**(-3)
```

```python
# Create a mask where the threshold is exceeded at any height
mask_4km = ash_insol_4km.max(dim='height') > 2
mask_9km = ash_insol_9km.max(dim='height') > 2

#print(mask_4km[48,:,:].values)
```

```python
# Map of no-fly areas
levels = [0.5,2]

crs_proj = ccrs.PlateCarree()

fig1 = plt.figure(figsize=(12,7))
ax1 = plt.subplot(2,2,1,projection=crs_proj)
c=plt.contourf(ds_4km.lon,ds_4km.lat,mask_4km[48,:,:],cmap='Reds',levels=levels,transform=crs_proj)
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-30, 30, 30, 65])
plt.title('No-fly zones for emission at 4km after 48h')

ax1 = plt.subplot(2,2,2,projection=crs_proj)
c=plt.contourf(ds_9km.lon,ds_9km.lat,mask_9km[48,:,:],cmap='Reds',levels=levels,transform=crs_proj)
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-30, 30, 30, 65])
plt.title('No-fly zones for emission at 9km after 48h')

ax1 = plt.subplot(2,2,3,projection=crs_proj)
c=plt.contourf(ds_4km.lon,ds_4km.lat,mask_4km[96,:,:],cmap='Reds',levels=levels,transform=crs_proj)
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-30, 30, 30, 65])
plt.title('No-fly zones for emission at 4km after 96h')

ax1 = plt.subplot(2,2,4,projection=crs_proj)
c=plt.contourf(ds_9km.lon,ds_9km.lat,mask_9km[96,:,:],cmap='Reds',levels=levels,transform=crs_proj)
ax1.coastlines(resolution='50m', linewidth=0.7)
ax1.set_extent([-30, 30, 30, 65])
plt.title('No-fly zones for emission at 9km after 96h')

plt.show()
```
